**Problem 8.** Bayes' rule â€” medical test example  

  - A disease has prevalence 1%. A diagnostic test has sensitivity 95% ($P(\text{ test+ } | \text{ disease })$)
    and specificity 90% ($P(\text{ test- } | \text{ no disease )}$).  

    (a) Compute $P(\text{ disease } | \text{ test+ })$. 

    **Answer:** 

    From Bayes' Law

    $$P(\text{ disease } | \text{ test+ }) 
      = \frac{P(\text{test+ } | \text{ disease }) P(\text{ disease })}{P(\text{ test+ })} \tag{8.1}$$

    where 

    $$P(\text{ test+ }) = P(\text{ test+ } | \text{ disease })P(\text{ disease }) 
                        + P(\text{ test+ } | \text{ no disease })P(\text{ no disease }) \tag{8.2}$$

    If the patient does not have the disease there are two possible outcomes: 
    a positive test or a negative test.  These are mutually exclusive and collectively
    exhaustive so

    $$P(\text{ test+ } | \text{ no disease }) + P(\text{ test- } | \text{ no disease }) = 1$$

    which rearranges to 

    $$P(\text{ test+ } | \text{ no disease }) = 1 - P(\text{ test- } | \text{ no disease }). \tag{8.3}$$

    Because a patient either has or does not have the disease and substituting (8.3) into (8.2), we get

    $$P(\text{ test+ }) = P(\text{ test+ } | \text{ disease })P(\text{ disease }) 
                        + (1 - P(\text{ test- } | \text{ no disease })) (1-P(\text{ disease })) \tag{8.5}$$

    We are given the prevalence, sensitivity, and specificity so (8.5) becomes

    $$P(\text{ test+ }) = 0.95 \cdot 0.01 + (1 - 0.9) \cdot (1 - 0.01) = 0.0095 + 0.099 = 0.1085 \tag{8.6}$$

    Substituting (8.6) into (8.1) yields

    $$P(\text{ disease } | \text{ test+ }) = \frac{0.95 \cdot 0.01}{0.1085} \approx \boxed{0.088}$$



(b) Compute $P(\text{ no disease } | \text{ test- })$.  

**Answer:** 

From Bayes' Law,

$$P(\text{ no disease } | \text{ test- } ) 
  = \frac{P( \text{ test- } | \text{ no disease } ) P( \text{ no disease })}{P(\text{ test- })} \tag{8.7}$$

From the Law of Total Probability

$$P(\text{ test- } ) = P(\text{ test- } | \text{ no disease }) P(\text{ no disease } ) 
                     + P(\text{ test- } | \text{ disease } ) P( \text{ disease }) 
                     = 0.9 \cdot 0.99 + (1-0.95) 0.01 = 0.8915 \tag{8.8}$$

Substituting (8.8) into (8.7) yields

$$P(\text{ no disease } | \text{ test- } ) = \frac{0.9 \cdot 0.99}{0.8915} \approx \boxed{0.9994}$$

**Problem 10.** Draw a 5-card hand uniformly at random from a standard 52-card deck (without replacement).


(a) Compute the probability the hand contains exactly two pairs (two distinct
    ranks each appearing twice, plus a fifth card of a different rank). Give the
    answer in terms of binomial coefficients (e.g., $n$ choose $k$).


**Answer**: 

Choose two numbers or faces (ranks) to be the pairs.   Then choose two suits for
each pair.  The remainnig card is any card from any suit so long as its face
or number does not match either pair.

$$P(\text{2 pairs}) = \frac{\binom{13}{2}\binom{4}{2}\binom{4}{2} \binom{4}{1} \cdot 11}{\binom{52}{5}} \approx \boxed{0.0475}$$


(b) Compute the probability the hand contains at least one ace. Express
    using complementary counting and combinatorial terms.

**Answer**:

$$P(\text{ at least 1 ace }) = 1 - P(\text{no aces}) = 1 - \frac{\binom{48}{5}}{\binom{52}{5}} \approx \boxed{0.341}$$




(c) Given the hand contains at least one ace, compute the conditional
    probability that it contains exactly two aces.

**Answer**:

If it has one ace then there are 3 other aces in the deck.  We want the case
when there are exactly 2.  This the probability of drawing exactly 1
ace with 4 draws when 1 has been removed from the deck.

I originally gave the answer below:

$$P(\text{ exactly 2 aces | at least 1 ace }) = \frac{\binom{3}{1}\binom{48}{3}}{\binom{48}{4}} = \frac{12}{45} \approx \boxed{0.267}$$

However, the denominator is missing certain cases.  Once we remove 1 ace, there are 51 cards remaining, not 48.

$$P(\text{ exactly 2 aces | at least 1 ace }) = \frac{\binom{3}{1}\binom{48}{3}}{\binom{51}{4}}  \approx \boxed{0.207}$$


```python
import random

# Function to simulate drawing 4 cards from a deck with 1 ace already in hand
def simulate_draws(num_simulations=20000):
    count_two_aces = 0

    for _ in range(num_simulations):
        # Create a deck with 51 cards (3 aces and 48 non-aces)
        deck = ['Ace'] * 3 + ['Non-Ace'] * 48
        random.shuffle(deck)

        # Draw 4 cards
        drawn_cards = random.sample(deck, 4)

        # Check if exactly 1 more ace is drawn (total 2 aces including the one in hand)
        if drawn_cards.count('Ace') == 1:
            count_two_aces += 1

    # Probability of having exactly 2 aces
    probability = count_two_aces / num_simulations
    return probability

# Run the simulation
probability_two_aces = simulate_draws()
print(f"Probability of having exactly 2 aces: {probability_two_aces:.4f}")
```

The sigmoid function is often used in machine learning and statistics to model probabilities. It is defined as:

$$
  \sigma(x) = \frac{1}{1 + e^{-x}}
$$

(a) Compute the value of $\sigma(0)$ and interpret its meaning in terms of probability.

**Answer 15(a)**

$$\sigma(0) = \frac{1}{1 + e^0} = \frac12$$

If this a CDF then it means that 1/2 of the probability is below zero and 1/2 above.

Althought this detail is not required for this answer, sigmoids are often used
as an activation function for a neural network node where values above 0 represent
activation and below zero represents deactivation, but with a smooth transition
so that the gradient is defined everywhere.



(b) Show that $\sigma(-x) = 1 - \sigma(x)$.
    
**Answer 15(b)**

\begin{align*}
  \sigma(-x) &= \frac{1}{1+ e^{-(-x)}} \\
             &= \frac{1}{1+e^x} \cdot \frac{e^{-x}}{e^{-x}} \\
             &= \frac{e^{-x}}{1+e^{-x}} \\
             &= \frac{1 - 1 + e^{-x}}{1+e^{-x}} \\
             &= 1 - \frac{1}{1 + e^{-x}} \\
             &= 1 - \sigma(x)
\end{align*}



(c) Plot the sigmoid function for $x \in [-10, 10]$ and describe its key properties (e.g., range, asymptotes, etc.).

**Answer 15(c)**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for x
x = np.linspace(-10, 10, 500)  # from -10 to 10 with 500 points
y = 1 / (1 + np.exp(-x))

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="$\sigma(x)$", linewidth=2)

# Add labels and title
plt.title("Plot of $\sigma(x)$")
plt.xlabel("$x$ (radians)")
plt.ylabel("$\sigma(x)$")

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

**Problem 16**. Hyperbolic Tangent Function

The hyperbolic tangent function, $\tanh(x)$, is another function commonly used in
machine learning. It is defined as:

$$
  \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

(a) Compute the value of $\tanh(0)$.

**Answer 16(a)**

$$\tanh(x) = \frac{e^0-e^0}{e^0 + e^0} = \frac{0}{2} = \boxed{0}$$


(b) Show that $\tanh(-x) = -\tanh(x)$.

**Answer 16(b)**

$$\tanh(-x)= \frac{e^{(-x)} - e^{-(-x)}}{e^{(-x)} + e^{-(-x)}} = - \frac{e^x - e^{-x}}{e^x + e^{-x}} = -\tanh(x)$$


(c) Plot the $\tanh(x)$ function for $x \in [-10, 10]$ and describe its key properties (e.g., range, asymptotes, symmetry, etc.).


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for x
x = np.linspace(-10, 10, 500)  # from -10 to 10 with 500 points
y = np.tanh(x)

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="tanh$(x)$", linewidth=2)

# Add labels and title
plt.title("Plot of tanh$(x)$")
plt.xlabel("$x$ (radians)")
plt.ylabel("tanh$(x)$")

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```


(d) Compare $\tanh(x)$ with the sigmoid function $\sigma(x)$ in terms of output range.

**Possible Answer 16(c)**

The range of the sigmoid is from zero to 1.  The range of the tanh function is from -1 to 1.

*Optional detail:* Although the following detail is not necessary to answer the question, a sigmoid can only represent
an output in the range of zero to 1, but when the output is scaled by a constant it allows the
sigmoid to represent any value in 0 to an arbitrarily high constant.  When a sigmoid is applied
to the output of a linear combination as with linear regression, the output can be constrained
to match only positive values.  Additionally it introduces a nonlinearity which as we will show
later is necessary for building networks of neurons where neurons can be viewed as 
scaled switch that are either "on" or "off."

A tanh importantly represent either positive or negative numbers, and although it is
limited to -1 to 1, by multiplying by a constant we can represent arbitrarily large
numbers anywhere in $\mathbb{R}$.   This allows our network to be trained to match 
both positive and negative values, e.g., as occurs in regression problems that are
have both positive and negative observed values.
