```python
# Prompt:
# Open cat.jpg with PIL and display with matplotlib using pixel units on the axes

# Generated code by ChatGPT 5.0-mini (preview)

from PIL import Image
import matplotlib.pyplot as plt

img = Image.open('cat.jpg')  # path relative to the notebook
width, height = img.size

fig, ax = plt.subplots()
ax.imshow(img)
ax.set_xlim(0, width)
ax.set_ylim(height, 0)  # flip y-axis so (0,0) is top-left in pixel coordinates
ax.set_xlabel('x (pixels)')
ax.set_ylabel('y (pixels)')
ax.set_aspect('equal')
plt.show()
```

```python
# Prompt:
# Crop the image from cat.jpg to the rage from 250 to 1400 in x and 
# 200 to 1800 in y then render the resulting cropped image.

# Generated code by ChatGPT 5.0-mini (preview)

# Crop region x:250..1400, y:200..1800 (PIL box: left, upper, right, lower)
crop_box = (250, 200, 1400, 1800)
cropped = img.crop(crop_box)

# Display cropped image in a new figure
fig2, ax2 = plt.subplots()
ax2.imshow(cropped)
w2, h2 = cropped.size
ax2.set_xlim(0, w2)
ax2.set_ylim(h2, 0)  # make (0,0) top-left in pixel coordinates
ax2.set_xlabel('x (pixels)')
ax2.set_ylabel('y (pixels)')
ax2.set_aspect('equal')
plt.show()
```

```python
# prompt: 
# extract red channel from the cropped PIL image and render it with pixel axes

# generated code by ChatGPT 5.0-mini (preview)
r, g, b = cropped.split()

fig3, ax3 = plt.subplots()
ax3.imshow(r, cmap='gray', vmin=0, vmax=255)
ax3.set_xlim(0, w2)
ax3.set_ylim(h2, 0)  # top-left origin in pixel coordinates
ax3.set_xlabel('x (pixels)')
ax3.set_ylabel('y (pixels)')
ax3.set_title('Red channel (cropped)')
ax3.set_aspect('equal')
plt.show()
```
